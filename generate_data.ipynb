```python
%env OPENAI_API_KEY= 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

#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

from openai import OpenAI
from PIL import Image
# from dotenv import load_dotenv
from dotenv import dotenv_values
config = dotenv_values('../.env')


def generate_dataset(num):

    # prompt = """
    #         You are an expert at extracting useful information from user queries. I need you to generate one information pair based on the user's query. 
    #         Users can be from various domains, and their queries aim to retrieve relevant files based on their needs. The extracted information should exclusively contain key-value pairs.
    #         Additionally, please generate 5 synonyms for the extracted 'file content'. Below are 5 examples that meet these requirements:

    #         Example1
    #         ### query: Show me budget reports from the second quarter of this year
    #         ### information: {'year':[0,0], 'month':[4, 6], 'day':[0, 0], 'file content':['Budget Report', 'Financial Statement', 'Budget Overview', 'Financial Report', 'Budget Summary'], 'file type':['pdf', 'xlsx', 'xls', 'docx', 'doc', 'csv', 'ppt', 'pptx', 'txt']}

    #         Example2
    #         ### query: Find my photos from New York last summer
    #         ### information: {'year': [-1, -1], 'month': [6, 8], 'day': [0, 0], 'file content': ['Photo taken in New York', 'New York Image', 'New York Snapshot', 'New York Picture', 'New York Photograph'], 'file type': ['jpg', 'jpeg', 'png', 'heif', 'tiff']}

    #         Example3
    #         ### query: Recent presentations on climate change
    #         ### information: {'year': [0, 0], 'month': [0, 0], 'day': [0, 0], 'file content': ['Climate Change Presentation', 'Global Warming Slides', 'Climate Crisis Deck', 'Environmental Change Presentation', 'Climate Action Slides'], 'file type': ['pdf', 'pptx', 'ppt', 'mp4', 'mov']}

    #         Example4
    #         ### query: Feedback documents received from design team on Project X
    #         ### information: {'year': [0, 0], 'month': [0, 0], 'day': [0, 0], 'file content': ['Design Team Feedback', 'Project X Review', 'Design Critique', 'Project X Feedback', 'Design Evaluation'], 'file type': ['pdf', 'docx', 'doc', 'txt', 'xlsx']}
            
    #         Example5
    #         ### query: Notes from my meetings with project team Alpha during last 3 months.
    #         ### information: {'year': [0, 0], 'month': [-3, 0], 'day': [0, 0], 'file content': ['Team Alpha Meeting Notes', 'Project Alpha Session Recap', 'Alpha Team Discussions', 'Meeting Outcomes with Team Alpha', 'Alpha Project Meeting Summaries'], 'file type': ['doc', 'docx', 'pdf', 'txt', 'md']}
    #         """

    prompt = """
            You are an expert at extracting useful information from user queries. I need you to generate one information pair based on the user's query. 
            Users can be from various domains, and their queries aim to retrieve relevant files based on their needs with a very specific date. DO NOT USE QUARTER TO DESCRIBE DATE.  
            The extracted information should exclusively contain key-value pairs. Additionally, please generate 5 synonyms for the extracted 'file content'. Below are 5 examples that meet these requirements:

            Example1
            ### query: Project documentation from January 15, 2024, to February 20, 2024
            ### information: {'year': [2024, 2024], 'month': [1, 2], 'day': [15, 20], 'file content': ['Project Documentation', 'Project Files', 'Project Overview', 'Project Details', 'Project Progress Documentation'], 'file type': ['pdf', 'doc', 'docx']}

            Example2
            ### query: Handouts from the digital marketing workshop series in February 2024
            ### information: {'year': [2024, 2024], 'month': [2, 2], 'day': [1, 28], 'file content': ['Digital Marketing Workshop Handouts', 'Marketing Workshop Materials', 'Digital Marketing Series Docs', 'Workshop Handouts', 'Digital Marketing Resources'], 'file type': ['pdf', 'pptx', 'doc', 'docx']}

            Example3
            ### query: Conference materials from the Global Tech Summit held from 2023/10/10 to 2023/10/12
            ### information: {'year': [2023, 2023], 'month': [10, 10], 'day': [10, 12], 'file content' : ['Global Tech Summit Materials', 'Tech Summit Presentations', 'Tech Conference Docs', 'Tech Summit Slides', 'Tech Summit Proceedings'], 'file type': ['pdf', 'pptx', 'doc', 'docx']}

            Example4
            ### query: Customer feedback collected in the week following March 10, 2024
            ### information: {'year': [2024, 2024], 'month': [3, 3], 'day': [11, 17], 'file content': ['Post-Launch Customer Feedback', 'Customer Responses Week After Launch', 'Feedback Collection Week Following Launch', 'Launch Week Customer Insights', 'Early Customer Reactions'], 'file type': ['txt', 'pdf', 'docx']}

            Example5
            ### query: Review the marketing strategy documents updated on 2023/10/12
            ### information: {'year': [2023, 2023], 'month': [10, 10], 'day': [12, 12], 'file content': ['Marketing Strategy Update', 'Updated Marketing Plans', 'Revised Marketing Strategy Documents', 'Marketing Strategy Revision', 'Marketing Plan Updates'], 'file type': ['pdf', 'doc', 'docx', 'pptx']}
            """

    # client = OpenAI(api_key=config['OPENAI_API_KEY'])
    client = OpenAI()
    response = client.chat.completions.create(
        model='gpt-4', #gpt-4
        messages=[
            {
                'role': 'user',
                'content': [
                    {'type': 'text', 'text': prompt},
                ],
            }
        ],
        max_tokens=2000,
    )
    query_tasks = response.choices[0].message.content
    
    return query_tasks

response = []
for i in range(200):
    print(len(response))
    response.append(generate_dataset(1))
```
