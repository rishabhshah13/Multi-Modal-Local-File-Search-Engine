```python
%env OPENAI_API_KEY= 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

from openai import OpenAI
from PIL import Image
# from dotenv import load_dotenv
from dotenv import dotenv_values
config = dotenv_values('../.env')


def generate_dataset(num):

    prompt = """
            You are an expert at extracting useful information from user queries. I need you to generate one information pair based on the user's query. 
            Users can be from various domains, and their queries aim to retrieve relevant files based on their needs. The extracted information should exclusively contain key-value pairs.
            Additionally, please generate 5 synonyms for the extracted 'file content'. Below are 5 examples that meet these requirements:

            Example1
            Task
            ### query: Show me budget reports from the second quarter of this year
            ### information: {'year':[0,0], 
                              'month':[4, 6], 
                              'day':[0, 0], 
                              'file content':['Budget Report', 'Financial Statement', 'Budget Overview', 'Financial Report', 'Budget Summary'], 
                              'file type':['pdf', 'xlsx', 'xls', 'docx', 'doc', 'csv', 'ppt', 'pptx', 'txt']}

            Example2
            Task
            ### query: Find my photos from New York last summer
            ### information: {'year': [-1, -1],
                              'month': [6, 8],
                              'day': [0, 0],
                              'file content': ['Photo taken in New York', 'New York Image', 'New York Snapshot', 'New York Picture', 'New York Photograph'],
                              'file type': ['jpg', 'jpeg', 'png', 'heif', 'tiff']}

            Example3
            Task
            ### query: Recent presentations on climate change
            ### information: {'year': [0, 0],
                              'month': [0, 0],
                              'day': [0, 0],
                              'file content': ['Climate Change Presentation', 'Global Warming Slides', 'Climate Crisis Deck', 'Environmental Change Presentation', 'Climate Action Slides'],
                              'file type': ['pdf', 'pptx', 'ppt', , "mp4", "mov"]}

            Example4
            Task
            ### query: Feedback documents received from design team on Project X
            ### information: {'year': [0, 0],
                              'month': [0, 0],
                              'day': [0, 0],
                              'file content': ['Design Team Feedback', 'Project X Review', 'Design Critique', 'Project X Feedback', 'Design Evaluation'],
                              'file type': ['pdf', 'docx', 'doc', 'txt', 'xlsx']}
            Example5
            Task
            ### query: Notes from my meetings with project team Alpha during last 3 months.
            ### information: {'year': [0, 0],
                              'month': [-3, 0],
                              'day': [0, 0],
                              'file content': ['Team Alpha Meeting Notes', 'Project Alpha Session Recap', 'Alpha Team Discussions', 'Meeting Outcomes with Team Alpha', 'Alpha Project Meeting Summaries'],
                              'file type': ['doc', 'docx', 'pdf', 'txt', 'md']}
            """

    # client = OpenAI(api_key=config['OPENAI_API_KEY'])
    client = OpenAI()
    response = client.chat.completions.create(
        model='gpt-4', #gpt-4
        messages=[
            {
                'role': 'user',
                'content': [
                    {'type': 'text', 'text': prompt},
                ],
            }
        ],
        max_tokens=2000,
    )
    query_tasks = response.choices[0].message.content
    
    return query_tasks

response = []
for i in range(1000):
    response.append(generate_dataset(1))
```

```python
import json
import re

# Given string
input_string = "Task\n### query: Emails regarding my application to Harvard Law School\n### information: {'year': [0, 0],\n                  'month': [0, 0],\n                  'day': [0, 0],\n                  'file content': ['Harvard Law School Application Email', 'Harvard Application Communication', 'Harvard Law Correspondence', 'Harvard Law School Application Conversations', 'Law School Application Email'],\n                  'file type': ['eml', 'msg', 'pdf', 'doc', 'txt']}"


data_list = []

for input_string in response:
    # Extracting relevant information
    query_start = input_string.find("query:")
    query_end = input_string.find("\n### information:")
    query = input_string[query_start + len("query:"):query_end].strip()
    info_start = input_string.find("{")
    info_end = input_string.rfind("}") + 1
    info = input_string[info_start:info_end]

    # Remove spaces between dictionary items while preserving spaces within list items
    info = re.sub(r',\s+', ', ', info)
    # Creating dictionary
    data = {"input": query, "output": info}
    data_list.append(data)


# Saving all dictionaries to a single JSON file
with open("query_finetune.json", "w") as json_file:
    json.dump(data_list, json_file, indent=4)
```
